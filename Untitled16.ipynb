```python
# Step 2: MIDI to Piano Roll Preprocessing WITH EMOTION LABELS
# Run this after Step 1

import muspy
import numpy as np
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt

print("=" * 60)
print("STEP 2: MIDI TO PIANO ROLL WITH EMOTION LABELS")
print("=" * 60)

# Configuration
BARS = 4  # Generate 4 bars of music
BEATS_PER_BAR = 4
TIME_STEPS_PER_BEAT = 4  # 16th note resolution
TOTAL_TIME_STEPS = BARS * BEATS_PER_BAR * TIME_STEPS_PER_BEAT  # 64 steps
PITCH_RANGE = 128  # Full MIDI range (0-127)

print(f"\nConfiguration:")
print(f"  Bars: {BARS}")
print(f"  Time steps: {TOTAL_TIME_STEPS}")
print(f"  Pitch range: {PITCH_RANGE}")
print(f"  Resolution: 16th notes")

# Load dataset (already downloaded)
emopia = muspy.EMOPIADataset("data/emopia", download_and_extract=True)
print(f"\nDataset size: {len(emopia)} tracks")

# Load emotion labels from Step 1
index_to_emotion = np.load('index_to_emotion.npy', allow_pickle=True).item()
print(f"Loaded emotion labels: {len(index_to_emotion)} tracks")

def music_to_piano_roll(music, total_steps=TOTAL_TIME_STEPS):
    """
    Convert MusPy Music object to piano roll.
    Returns: numpy array of shape (total_steps, 128)
    """
    # Initialize empty piano roll
    piano_roll = np.zeros((total_steps, PITCH_RANGE), dtype=np.float32)

    # Get all notes from all tracks
    all_notes = []
    for track in music.tracks:
        if track.notes:
            all_notes.extend(track.notes)

    if not all_notes:
        return piano_roll

    # Convert time to steps (assuming 4/4 time, resolution = ticks per quarter note)
    resolution = music.resolution if music.resolution else 480
    ticks_per_step = resolution / TIME_STEPS_PER_BEAT

    # Fill piano roll
    for note in all_notes:
        # Convert time to step index
        step_start = int(note.time / ticks_per_step)
        step_end = int((note.time + note.duration) / ticks_per_step)

        # Clip to valid range
        step_start = max(0, min(step_start, total_steps - 1))
        step_end = max(0, min(step_end, total_steps))

        # Set note on
        if step_start < total_steps and 0 <= note.pitch < PITCH_RANGE:
            for step in range(step_start, step_end):
                if step < total_steps:
                    piano_roll[step, note.pitch] = 1.0

    return piano_roll

print("\n" + "=" * 60)
print("PREPROCESSING ALL TRACKS WITH EMOTIONS...")
print("=" * 60)

# Process all tracks
piano_rolls = []
emotion_labels = []
valid_tracks = 0

for idx in tqdm(range(len(emopia)), desc="Converting MIDI to piano rolls"):
    try:
        # Only process tracks with emotion labels
        if idx not in index_to_emotion:
            continue

        music = emopia[idx]
        piano_roll = music_to_piano_roll(music)

        # Only keep piano rolls with some notes
        if piano_roll.sum() > 10:  # At least 10 active notes
            piano_rolls.append(piano_roll)
            emotion_labels.append(index_to_emotion[idx])
            valid_tracks += 1
    except Exception as e:
        # Skip problematic files
        continue

piano_rolls = np.array(piano_rolls)  # Shape: (N, 64, 128)
emotion_labels = np.array(emotion_labels)  # Shape: (N,)

print(f"\n✓ Preprocessing complete!")
print(f"  Valid tracks: {valid_tracks}")
print(f"  Piano roll shape: {piano_rolls.shape}")
print(f"  Emotion labels shape: {emotion_labels.shape}")
print(f"  Data range: [{piano_rolls.min():.2f}, {piano_rolls.max():.2f}]")
print(f"  Average density: {piano_rolls.mean():.4f} (ratio of active notes)")

# Emotion distribution
from collections import Counter
emotion_counts = Counter(emotion_labels)
print(f"\nEmotion Distribution in Preprocessed Data:")
for q in sorted(emotion_counts.keys()):
    print(f"  Q{q}: {emotion_counts[q]} tracks ({emotion_counts[q]/len(emotion_labels)*100:.1f}%)")

# Save preprocessed data
print("\nSaving preprocessed data...")
np.save('piano_rolls.npy', piano_rolls)
np.save('emotion_labels.npy', emotion_labels)
print("✓ Saved to piano_rolls.npy and emotion_labels.npy")

# Visualize samples from each emotion
print("\n" + "=" * 60)
print("VISUALIZATION: SAMPLE PIANO ROLLS PER EMOTION")
print("=" * 60)

emotion_names = {
    1: "Q1: Happy/Excited",
    2: "Q2: Angry/Tense",
    3: "Q3: Sad/Depressed",
    4: "Q4: Peaceful/Relaxed"
}

fig, axes = plt.subplots(2, 4, figsize=(16, 8))

for i, emotion in enumerate(sorted(emotion_counts.keys())):
    # Find first two samples of this emotion
    indices = np.where(emotion_labels == emotion)[0][:2]

    for j, idx in enumerate(indices):
        ax = axes[j, i]
        # Transpose to show pitch on Y-axis, time on X-axis
        ax.imshow(piano_rolls[idx].T, aspect='auto', cmap='Blues', origin='lower')
        ax.set_xlabel('Time Step', fontsize=9)
        ax.set_ylabel('MIDI Pitch', fontsize=9)
        ax.set_title(f'{emotion_names[emotion]}\nSample {j+1}', fontsize=10, fontweight='bold')
        ax.set_ylim([21, 108])  # Standard piano range

plt.tight_layout()
plt.savefig('piano_roll_samples_by_emotion.png', dpi=100, bbox_inches='tight')
plt.show()

print(f"\n✓ Saved visualization to piano_roll_samples_by_emotion.png")

# Statistics per emotion
print("\n" + "=" * 60)
print("DATASET STATISTICS PER EMOTION")
print("=" * 60)

for emotion in sorted(emotion_counts.keys()):
    indices = np.where(emotion_labels == emotion)[0]
    emotion_piano_rolls = piano_rolls[indices]

    note_counts = emotion_piano_rolls.sum(axis=(1, 2))
    pitch_usage = emotion_piano_rolls.sum(axis=(0, 1))
    active_pitches = np.where(pitch_usage > 0)[0]

    print(f"\n{emotion_names[emotion]}:")
    print(f"  Tracks: {len(indices)}")
    print(f"  Avg notes per track: {note_counts.mean():.1f} ± {note_counts.std():.1f}")
    print(f"  Pitch range: {active_pitches.min()}-{active_pitches.max()} MIDI")
    print(f"  Note density: {emotion_piano_rolls.mean():.4f}")

print("\n" + "=" * 60)
print("✓ STEP 2 COMPLETE - DATA PREPROCESSED WITH EMOTIONS!")
print("=" * 60)
print(f"\nReady for training with {len(piano_rolls)} labeled samples")
print("Next: Run Step 3 to build and train the EMOTION-CONDITIONED VAE")
```

```python
# Step 3: EMOTION-CONDITIONED VAE Architecture & Training
# Run this after Step 2

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

print("=" * 60)
print("STEP 3: EMOTION-CONDITIONED VAE TRAINING")
print("=" * 60)

# Check GPU availability
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nDevice: {device}")

# Load preprocessed data
piano_rolls = np.load('piano_rolls.npy')
emotion_labels = np.load('emotion_labels.npy')
print(f"Loaded data: {piano_rolls.shape}")
print(f"Loaded labels: {emotion_labels.shape}")

# Configuration
NUM_EMOTIONS = 4  # Q1, Q2, Q3, Q4
EMOTION_EMBED_DIM = 64
LATENT_DIM = 512
HIDDEN_DIM = 1024
BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 1e-3
BETA = 0.1  # Weight for KL divergence

print(f"\nEmotion-Conditioned VAE Configuration:")
print(f"  Number of emotions: {NUM_EMOTIONS}")
print(f"  Emotion embedding dim: {EMOTION_EMBED_DIM}")
print(f"  Latent dimension: {LATENT_DIM}")
print(f"  Hidden dimension: {HIDDEN_DIM}")
print(f"  Batch size: {BATCH_SIZE}")
print(f"  Epochs: {EPOCHS}")
print(f"  Learning rate: {LEARNING_RATE}")
print(f"  Beta (KL weight): {BETA}")

# Dataset class
class EmotionPianoRollDataset(Dataset):
    def __init__(self, piano_rolls, emotion_labels):
        self.data = torch.FloatTensor(piano_rolls)
        # Convert emotions from 1-4 to 0-3 for embedding
        self.emotions = torch.LongTensor(emotion_labels - 1)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.emotions[idx]

# Create dataset and dataloader
dataset = EmotionPianoRollDataset(piano_rolls, emotion_labels)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)

print(f"\nDataset size: {len(dataset)} samples")
print(f"Batches per epoch: {len(dataloader)}")

# Emotion-Conditioned VAE Model
class EmotionConditionedVAE(nn.Module):
    def __init__(self, input_dim=64*128, hidden_dim=HIDDEN_DIM, latent_dim=LATENT_DIM,
                 num_emotions=NUM_EMOTIONS, emotion_embed_dim=EMOTION_EMBED_DIM):
        super(EmotionConditionedVAE, self).__init__()

        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.emotion_embed_dim = emotion_embed_dim

        # Emotion embedding layer
        self.emotion_embedding = nn.Embedding(num_emotions, emotion_embed_dim)

        # Encoder (takes piano roll + emotion embedding)
        self.encoder = nn.Sequential(
            nn.Linear(input_dim + emotion_embed_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2)
        )

        # Latent space
        self.fc_mu = nn.Linear(hidden_dim // 2, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim // 2, latent_dim)

        # Decoder (takes latent + emotion embedding)
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim + emotion_embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 2, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()  # Output in [0, 1]
        )

    def encode(self, x, emotion):
        # Flatten piano roll: (B, 64, 128) -> (B, 8192)
        x = x.view(-1, self.input_dim)

        # Get emotion embedding
        emotion_emb = self.emotion_embedding(emotion)  # (B, emotion_embed_dim)

        # Concatenate input with emotion
        x_cond = torch.cat([x, emotion_emb], dim=1)

        # Encode
        h = self.encoder(x_cond)
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z, emotion):
        # Get emotion embedding
        emotion_emb = self.emotion_embedding(emotion)  # (B, emotion_embed_dim)

        # Concatenate latent with emotion
        z_cond = torch.cat([z, emotion_emb], dim=1)

        # Decode
        x = self.decoder(z_cond)
        # Reshape: (B, 8192) -> (B, 64, 128)
        return x.view(-1, 64, 128)

    def forward(self, x, emotion):
        mu, logvar = self.encode(x, emotion)
        z = self.reparameterize(mu, logvar)
        recon = self.decode(z, emotion)
        return recon, mu, logvar

# Loss function
def vae_loss(recon_x, x, mu, logvar, beta=BETA):
    # Reconstruction loss (Binary Cross Entropy)
    BCE = F.binary_cross_entropy(recon_x, x, reduction='sum')

    # KL Divergence
    KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

    return BCE + beta * KLD, BCE, KLD

# Initialize model
model = EmotionConditionedVAE().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

print("\n" + "=" * 60)
print("EMOTION-CONDITIONED VAE ARCHITECTURE")
print("=" * 60)
print(model)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")

# Training loop
print("\n" + "=" * 60)
print("TRAINING EMOTION-CONDITIONED VAE")
print("=" * 60)

train_losses = []
recon_losses = []
kl_losses = []

for epoch in range(EPOCHS):
    model.train()
    epoch_loss = 0
    epoch_recon = 0
    epoch_kl = 0

    pbar = tqdm(dataloader, desc=f'Epoch {epoch+1}/{EPOCHS}')
    for batch, emotions in pbar:
        batch = batch.to(device)
        emotions = emotions.to(device)

        # Forward pass
        recon, mu, logvar = model(batch, emotions)
        loss, recon_loss, kl_loss = vae_loss(recon, batch, mu, logvar)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Track losses
        epoch_loss += loss.item()
        epoch_recon += recon_loss.item()
        epoch_kl += kl_loss.item()

        pbar.set_postfix({
            'Loss': f'{loss.item()/len(batch):.2f}',
            'Recon': f'{recon_loss.item()/len(batch):.2f}',
            'KL': f'{kl_loss.item()/len(batch):.2f}'
        })

    # Average losses
    avg_loss = epoch_loss / len(dataset)
    avg_recon = epoch_recon / len(dataset)
    avg_kl = epoch_kl / len(dataset)

    train_losses.append(avg_loss)
    recon_losses.append(avg_recon)
    kl_losses.append(avg_kl)

    print(f'Epoch {epoch+1}: Loss={avg_loss:.2f}, Recon={avg_recon:.2f}, KL={avg_kl:.2f}')

print("\n✓ Training complete!")

# Save model
torch.save(model.state_dict(), 'emotion_vae_model.pth')
print("✓ Model saved to emotion_vae_model.pth")

# Plot training curves
print("\n" + "=" * 60)
print("TRAINING CURVES")
print("=" * 60)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(train_losses, linewidth=2)
axes[0].set_title('Total Loss', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].grid(alpha=0.3)

axes[1].plot(recon_losses, color='orange', linewidth=2)
axes[1].set_title('Reconstruction Loss', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].grid(alpha=0.3)

axes[2].plot(kl_losses, color='green', linewidth=2)
axes[2].set_title('KL Divergence', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Loss')
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('emotion_vae_training_curves.png', dpi=100, bbox_inches='tight')
plt.show()

print("✓ Saved training curves to emotion_vae_training_curves.png")

# Test reconstruction for each emotion
print("\n" + "=" * 60)
print("TESTING VAE RECONSTRUCTION PER EMOTION")
print("=" * 60)

emotion_names = {0: 'Q1: Happy', 1: 'Q2: Angry', 2: 'Q3: Sad', 3: 'Q4: Peaceful'}

model.eval()
with torch.no_grad():
    fig, axes = plt.subplots(4, 2, figsize=(12, 16))

    for emotion_idx in range(NUM_EMOTIONS):
        # Find a sample with this emotion
        sample_indices = np.where(emotion_labels == emotion_idx + 1)[0]
        if len(sample_indices) > 0:
            test_idx = sample_indices[0]
            test_sample = dataset[test_idx][0].unsqueeze(0).to(device)
            test_emotion = dataset[test_idx][1].unsqueeze(0).to(device)

            recon, mu, logvar = model(test_sample, test_emotion)

            # Original
            axes[emotion_idx, 0].imshow(test_sample[0].cpu().numpy().T, aspect='auto',
                                       cmap='Blues', origin='lower')
            axes[emotion_idx, 0].set_title(f'{emotion_names[emotion_idx]} - Original',
                                          fontsize=11, fontweight='bold')
            axes[emotion_idx, 0].set_xlabel('Time Step')
            axes[emotion_idx, 0].set_ylabel('MIDI Pitch')
            axes[emotion_idx, 0].set_ylim([21, 108])

            # Reconstructed
            axes[emotion_idx, 1].imshow(recon[0].cpu().numpy().T, aspect='auto',
                                       cmap='Blues', origin='lower')
            axes[emotion_idx, 1].set_title(f'{emotion_names[emotion_idx]} - Reconstructed',
                                          fontsize=11, fontweight='bold')
            axes[emotion_idx, 1].set_xlabel('Time Step')
            axes[emotion_idx, 1].set_ylabel('MIDI Pitch')
            axes[emotion_idx, 1].set_ylim([21, 108])

    plt.tight_layout()
    plt.savefig('emotion_vae_reconstruction.png', dpi=100, bbox_inches='tight')
    plt.show()

print("✓ Saved reconstruction comparison to emotion_vae_reconstruction.png")

print("\n" + "=" * 60)
print("✓ STEP 3 COMPLETE - EMOTION-CONDITIONED VAE TRAINED!")
print("=" * 60)
print("\nNext: Run Step 4 to build and train the EMOTION-CONDITIONED GAN")
```

```python
# Step 4: EMOTION-CONDITIONED GAN Training
# Run this after Step 3

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

print("=" * 60)
print("STEP 4: EMOTION-CONDITIONED GAN TRAINING")
print("=" * 60)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nDevice: {device}")

# Define Emotion-Conditioned VAE (same as Step 3)
class EmotionConditionedVAE(nn.Module):
    def __init__(self, input_dim=64*128, hidden_dim=1024, latent_dim=512,
                 num_emotions=4, emotion_embed_dim=64):
        super(EmotionConditionedVAE, self).__init__()

        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.emotion_embed_dim = emotion_embed_dim

        self.emotion_embedding = nn.Embedding(num_emotions, emotion_embed_dim)

        self.encoder = nn.Sequential(
            nn.Linear(input_dim + emotion_embed_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2)
        )

        self.fc_mu = nn.Linear(hidden_dim // 2, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim // 2, latent_dim)

        self.decoder = nn.Sequential(
            nn.Linear(latent_dim + emotion_embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 2, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()
        )

    def encode(self, x, emotion):
        x = x.view(-1, self.input_dim)
        emotion_emb = self.emotion_embedding(emotion)
        x_cond = torch.cat([x, emotion_emb], dim=1)
        h = self.encoder(x_cond)
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

    def decode(self, z, emotion):
        emotion_emb = self.emotion_embedding(emotion)
        z_cond = torch.cat([z, emotion_emb], dim=1)
        x = self.decoder(z_cond)
        return x.view(-1, 64, 128)

# Load trained VAE
vae = EmotionConditionedVAE().to(device)
vae.load_state_dict(torch.load('emotion_vae_model.pth', map_location=device))
vae.eval()
print("✓ Emotion-Conditioned VAE model loaded")

# Load data
piano_rolls = np.load('piano_rolls.npy')
emotion_labels = np.load('emotion_labels.npy')
dataset = torch.FloatTensor(piano_rolls)
emotions_tensor = torch.LongTensor(emotion_labels - 1)  # Convert to 0-3

print(f"Dataset size: {len(dataset)} samples")

# Extract latent vectors with emotions
print("\nExtracting latent vectors from Emotion-Conditioned VAE...")
latent_vectors = []
latent_emotions = []
batch_size = 64

with torch.no_grad():
    for i in tqdm(range(0, len(dataset), batch_size)):
        batch = dataset[i:i+batch_size].to(device)
        batch_emotions = emotions_tensor[i:i+batch_size].to(device)

        mu, _ = vae.encode(batch, batch_emotions)
        latent_vectors.append(mu.cpu())
        latent_emotions.append(batch_emotions.cpu())

latent_vectors = torch.cat(latent_vectors, dim=0)
latent_emotions = torch.cat(latent_emotions, dim=0)
print(f"✓ Extracted latent vectors: {latent_vectors.shape}")
print(f"✓ Extracted emotions: {latent_emotions.shape}")

# Normalize latent vectors
latent_mean = latent_vectors.mean(dim=0)
latent_std = latent_vectors.std(dim=0)
latent_vectors_normalized = (latent_vectors - latent_mean) / (latent_std + 1e-8)

print(f"✓ Normalized latent vectors")
print(f"  Mean: {latent_vectors_normalized.mean():.4f}")
print(f"  Std: {latent_vectors_normalized.std():.4f}")

torch.save({'mean': latent_mean, 'std': latent_std}, 'emotion_latent_stats.pth')

# GAN Configuration
NOISE_DIM = 128
EMOTION_EMBED_DIM = 64
LATENT_DIM = 512
HIDDEN_DIM = 512
BATCH_SIZE = 64
EPOCHS = 100
LR_G = 3e-4
LR_D = 1e-5
LABEL_SMOOTH_REAL = 0.9
LABEL_SMOOTH_FAKE = 0.1
N_CRITIC = 1
N_GEN = 5

print(f"\nEmotion-Conditioned GAN Configuration:")
print(f"  Noise dimension: {NOISE_DIM}")
print(f"  Emotion embedding dim: {EMOTION_EMBED_DIM}")
print(f"  Latent dimension: {LATENT_DIM}")
print(f"  Batch size: {BATCH_SIZE}")
print(f"  Epochs: {EPOCHS}")
print(f"  Generator LR: {LR_G}")
print(f"  Discriminator LR: {LR_D}")
print(f"  G updates per D update: {N_GEN}")

# Emotion-Conditioned Generator
class EmotionConditionedGenerator(nn.Module):
    def __init__(self, noise_dim=NOISE_DIM, latent_dim=LATENT_DIM,
                 hidden_dim=HIDDEN_DIM, num_emotions=4, emotion_embed_dim=EMOTION_EMBED_DIM):
        super(EmotionConditionedGenerator, self).__init__()

        self.emotion_embedding = nn.Embedding(num_emotions, emotion_embed_dim)

        self.model = nn.Sequential(
            nn.Linear(noise_dim + emotion_embed_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.BatchNorm1d(hidden_dim),

            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.BatchNorm1d(hidden_dim),

            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.BatchNorm1d(hidden_dim),

            nn.Linear(hidden_dim, latent_dim)
        )

    def forward(self, noise, emotion):
        emotion_emb = self.emotion_embedding(emotion)
        x = torch.cat([noise, emotion_emb], dim=1)
        return self.model(x)

# Emotion-Conditioned Discriminator
class EmotionConditionedDiscriminator(nn.Module):
    def __init__(self, latent_dim=LATENT_DIM, hidden_dim=HIDDEN_DIM,
                 num_emotions=4, emotion_embed_dim=EMOTION_EMBED_DIM):
        super(EmotionConditionedDiscriminator, self).__init__()

        self.emotion_embedding = nn.Embedding(num_emotions, emotion_embed_dim)

        self.model = nn.Sequential(
            nn.Linear(latent_dim + emotion_embed_dim, hidden_dim // 2),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),

            nn.Linear(hidden_dim // 2, hidden_dim // 4),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),

            nn.Linear(hidden_dim // 4, 1),
            nn.Sigmoid()
        )

    def forward(self, latent, emotion):
        emotion_emb = self.emotion_embedding(emotion)
        x = torch.cat([latent, emotion_emb], dim=1)
        return self.model(x)

# Initialize models
generator = EmotionConditionedGenerator().to(device)
discriminator = EmotionConditionedDiscriminator().to(device)

def weights_init(m):
    if isinstance(m, nn.Linear):
        nn.init.normal_(m.weight, 0.0, 0.02)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0)
    elif isinstance(m, nn.BatchNorm1d):
        nn.init.normal_(m.weight, 1.0, 0.02)
        nn.init.constant_(m.bias, 0)

generator.apply(weights_init)
discriminator.apply(weights_init)

optimizer_G = optim.Adam(generator.parameters(), lr=LR_G, betas=(0.5, 0.999))
optimizer_D = optim.Adam(discriminator.parameters(), lr=LR_D, betas=(0.5, 0.999))

criterion = nn.BCELoss()

print("\n" + "=" * 60)
print("EMOTION-CONDITIONED GENERATOR ARCHITECTURE")
print("=" * 60)
print(generator)
print(f"Parameters: {sum(p.numel() for p in generator.parameters()):,}")

print("\n" + "=" * 60)
print("EMOTION-CONDITIONED DISCRIMINATOR ARCHITECTURE")
print("=" * 60)
print(discriminator)
print(f"Parameters: {sum(p.numel() for p in discriminator.parameters()):,}")

# Create dataloader
latent_dataset = TensorDataset(latent_vectors_normalized, latent_emotions)
latent_loader = DataLoader(latent_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0)

# Training loop
print("\n" + "=" * 60)
print("TRAINING EMOTION-CONDITIONED GAN")
print("=" * 60)

g_losses = []
d_losses = []
d_real_accs = []
d_fake_accs = []

for epoch in range(EPOCHS):
    generator.train()
    discriminator.train()

    epoch_g_loss = 0
    epoch_d_loss = 0
    epoch_d_real_acc = 0
    epoch_d_fake_acc = 0

    pbar = tqdm(latent_loader, desc=f'Epoch {epoch+1}/{EPOCHS}')

    for batch_idx, (real_latent, emotions) in enumerate(pbar):
        batch_size_current = real_latent.size(0)
        real_latent = real_latent.to(device)
        emotions = emotions.to(device)

        real_labels = torch.empty(batch_size_current, 1).uniform_(
            LABEL_SMOOTH_REAL - 0.1, LABEL_SMOOTH_REAL + 0.1).to(device)
        fake_labels = torch.empty(batch_size_current, 1).uniform_(
            0.0, LABEL_SMOOTH_FAKE).to(device)

        # Train Discriminator
        if batch_idx % N_CRITIC == 0:
            optimizer_D.zero_grad()

            # Real samples
            real_output = discriminator(real_latent, emotions)
            d_real_loss = criterion(real_output, real_labels)

            # Fake samples
            noise = torch.randn(batch_size_current, NOISE_DIM).to(device)
            fake_latent = generator(noise, emotions)
            fake_output = discriminator(fake_latent.detach(), emotions)
            d_fake_loss = criterion(fake_output, fake_labels)

            d_loss = d_real_loss + d_fake_loss
            d_loss.backward()
            optimizer_D.step()

            epoch_d_loss += d_loss.item()
            epoch_d_real_acc += (real_output > 0.5).float().mean().item()
            epoch_d_fake_acc += (fake_output < 0.5).float().mean().item()

        # Train Generator
        for _ in range(N_GEN):
            optimizer_G.zero_grad()

            noise = torch.randn(batch_size_current, NOISE_DIM).to(device)
            fake_latent = generator(noise, emotions)
            fake_output = discriminator(fake_latent, emotions)

            g_loss = criterion(fake_output, real_labels)
            g_loss.backward()
            optimizer_G.step()

        epoch_g_loss += g_loss.item()

        pbar.set_postfix({
            'G_Loss': f'{g_loss.item():.3f}',
            'D_Loss': f'{d_loss.item():.3f}',
            'D_Real': f'{(real_output > 0.5).float().mean().item():.2f}',
            'D_Fake': f'{(fake_output > 0.5).float().mean().item():.2f}'
        })

    num_batches = len(latent_loader)
    g_losses.append(epoch_g_loss / (num_batches * N_GEN))
    d_losses.append(epoch_d_loss / (num_batches // N_CRITIC))
    d_real_accs.append(epoch_d_real_acc / (num_batches // N_CRITIC))
    d_fake_accs.append(epoch_d_fake_acc / (num_batches // N_CRITIC))

    print(f'Epoch {epoch+1}: G_Loss={g_losses[-1]:.3f}, D_Loss={d_losses[-1]:.3f}, '
          f'D_Real_Acc={d_real_accs[-1]:.2f}, D_Fake_Acc={d_fake_accs[-1]:.2f}')

print("\n✓ GAN Training complete!")

# Save models
torch.save(generator.state_dict(), 'emotion_generator.pth')
torch.save(discriminator.state_dict(), 'emotion_discriminator.pth')
print("✓ Models saved")

# Plot training curves
print("\n" + "=" * 60)
print("GAN TRAINING CURVES")
print("=" * 60)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(g_losses, label='Generator Loss', linewidth=2, color='blue')
axes[0].plot(d_losses, label='Discriminator Loss', linewidth=2, color='orange')
axes[0].set_title('Emotion-Conditioned GAN Losses', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(alpha=0.3)

axes[1].plot(d_real_accs, label='D Real Accuracy', linewidth=2, color='blue')
axes[1].plot(d_fake_accs, label='D Fake Accuracy', linewidth=2, color='orange')
axes[1].axhline(y=0.5, color='gray', linestyle='--', label='Random Guess', linewidth=2)
axes[1].set_title('Discriminator Accuracy', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_ylim([0, 1])
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('emotion_gan_training_curves.png', dpi=100, bbox_inches='tight')
plt.show()

print("✓ Saved training curves")
print("\n" + "=" * 60)
print("✓ STEP 4 COMPLETE - EMOTION-CONDITIONED GAN TRAINED!")
print("=" * 60)
print("\nNext: Run Step 5 to generate emotion-conditioned music!")
```

```python
# PRAGMATIC APPROACH: Ensemble + Data Augmentation for 61%+
# Strategy: Multiple models + better training tricks

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.model_selection import train_test_split, StratifiedKFold
import seaborn as sns
from collections import Counter
import pickle
import torch.nn.functional as F

print("=" * 70)
print("PRAGMATIC EMOPIA CLASSIFIER - ENSEMBLE APPROACH")
print("Strategy: Multiple architectures + heavy augmentation + ensembling")
print("=" * 70)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")

# Load data
piano_rolls = np.load('piano_rolls.npy')
emotion_labels = np.load('emotion_labels.npy')
print(f"\nLoaded: {piano_rolls.shape}, Labels: {emotion_labels.shape}")

# ============================================================================
# DATA AUGMENTATION
# ============================================================================

def augment_piano_roll(piano_roll):
    """Apply random augmentations"""
    aug_pr = piano_roll.copy()

    # Random pitch shift (-6 to +6 semitones)
    if np.random.rand() > 0.5:
        shift = np.random.randint(-6, 7)
        if shift > 0:
            aug_pr = np.pad(aug_pr[:, :-shift], ((0,0), (shift,0)), mode='constant')
        elif shift < 0:
            aug_pr = np.pad(aug_pr[:, -shift:], ((0,0), (0,-shift)), mode='constant')

    # Random time stretch (0.9x to 1.1x)
    if np.random.rand() > 0.5:
        from scipy.ndimage import zoom
        stretch_factor = np.random.uniform(0.9, 1.1)
        aug_pr = zoom(aug_pr, (stretch_factor, 1), order=1)
        if aug_pr.shape[0] > 64:
            aug_pr = aug_pr[:64]
        elif aug_pr.shape[0] < 64:
            aug_pr = np.pad(aug_pr, ((0, 64-aug_pr.shape[0]), (0, 0)), mode='constant')

    # Random velocity scaling (0.8x to 1.2x)
    if np.random.rand() > 0.5:
        scale = np.random.uniform(0.8, 1.2)
        aug_pr = np.clip(aug_pr * scale, 0, 1)

    return aug_pr

# ============================================================================
# MULTIPLE ARCHITECTURES
# ============================================================================

class CNNClassifier(nn.Module):
    """Pure CNN architecture"""
    def __init__(self, num_emotions=4):
        super(CNNClassifier, self).__init__()

        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=(3, 7), padding=(1, 3)),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d((2, 2)),
            nn.Dropout2d(0.3),

            nn.Conv2d(32, 64, kernel_size=(3, 7), padding=(1, 3)),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d((2, 2)),
            nn.Dropout2d(0.3),

            nn.Conv2d(64, 128, kernel_size=(3, 5), padding=(1, 2)),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 8))
        )

        self.fc = nn.Sequential(
            nn.Linear(128*4*8, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.BatchNorm1d(256),
            nn.Linear(256, num_emotions)
        )

    def forward(self, x):
        x = self.conv(x.unsqueeze(1))
        x = x.flatten(1)
        return self.fc(x)

class LSTMClassifier(nn.Module):
    """LSTM-based architecture"""
    def __init__(self, num_emotions=4):
        super(LSTMClassifier, self).__init__()

        self.lstm = nn.LSTM(
            input_size=128,
            hidden_size=256,
            num_layers=2,
            batch_first=True,
            bidirectional=True,
            dropout=0.3
        )

        self.attention = nn.Sequential(
            nn.Linear(512, 256),
            nn.Tanh(),
            nn.Linear(256, 1)
        )

        self.fc = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.BatchNorm1d(256),
            nn.Linear(256, num_emotions)
        )

    def forward(self, x):
        # x: (batch, 64, 128)
        lstm_out, _ = self.lstm(x)  # (batch, 64, 512)

        # Attention
        attn_weights = self.attention(lstm_out)  # (batch, 64, 1)
        attn_weights = F.softmax(attn_weights, dim=1)

        # Weighted sum
        context = torch.sum(lstm_out * attn_weights, dim=1)  # (batch, 512)

        return self.fc(context)

class ResNetClassifier(nn.Module):
    """ResNet-style architecture"""
    def __init__(self, num_emotions=4):
        super(ResNetClassifier, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, kernel_size=7, padding=3)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(2)

        # Residual blocks
        self.res1 = self._make_residual_block(32, 64)
        self.res2 = self._make_residual_block(64, 128)
        self.res3 = self._make_residual_block(128, 256)

        self.global_pool = nn.AdaptiveAvgPool2d((1, 1))

        self.fc = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_emotions)
        )

    def _make_residual_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(),
            nn.Conv2d(out_channels, out_channels, 3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.MaxPool2d(2)
        )

    def forward(self, x):
        x = x.unsqueeze(1)
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.pool1(x)

        x = self.res1(x)
        x = self.res2(x)
        x = self.res3(x)

        x = self.global_pool(x)
        x = x.flatten(1)
        return self.fc(x)

# ============================================================================
# DATASET
# ============================================================================

class AugmentedDataset(Dataset):
    def __init__(self, piano_rolls, labels, augment=False):
        self.piano_rolls = piano_rolls
        self.labels = labels - 1  # 0-3
        self.augment = augment

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        pr = self.piano_rolls[idx]

        if self.augment and np.random.rand() > 0.5:
            pr = augment_piano_roll(pr)

        return torch.FloatTensor(pr), torch.LongTensor([self.labels[idx]])[0]

# ============================================================================
# TRAINING UTILS
# ============================================================================

def train_model(model, train_loader, val_loader, epochs=80, lr=1e-3):
    """Train single model"""

    # Class weights
    emotion_counts = Counter(emotion_labels)
    class_weights = []
    for emotion in range(4):
        count = emotion_counts[emotion + 1]
        weight = len(emotion_labels) / (4 * count)
        class_weights.append(weight)
    class_weights = torch.FloatTensor(class_weights).to(device)

    criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.1)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2)

    best_val_acc = 0
    patience = 0

    for epoch in range(epochs):
        # Train
        model.train()
        train_loss = 0
        for data, labels in train_loader:
            data, labels = data.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            train_loss += loss.item()

        # Validate
        model.eval()
        correct = 0
        total = 0
        val_loss = 0
        with torch.no_grad():
            for data, labels in val_loader:
                data, labels = data.to(device), labels.to(device)
                outputs = model(data)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

        val_acc = 100. * correct / total
        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        scheduler.step()

        # Print every 5 epochs
        if (epoch + 1) % 5 == 0 or val_acc > best_val_acc:
            status = '✓ BEST' if val_acc > best_val_acc else ''
            print(f'Epoch {epoch+1:3d}/{epochs}: Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}% {status}')

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience = 0
        else:
            patience += 1

        if patience >= 15:
            print(f'Early stopping at epoch {epoch+1}')
            break

    return best_val_acc

# ============================================================================
# ENSEMBLE TRAINING
# ============================================================================

print("\n" + "=" * 70)
print("TRAINING ENSEMBLE OF 3 ARCHITECTURES")
print("=" * 70)

# Stratified split
indices = np.arange(len(piano_rolls))
labels = emotion_labels - 1

train_idx, temp_idx = train_test_split(indices, test_size=0.3, stratify=labels, random_state=42)
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, stratify=labels[temp_idx], random_state=42)

# Create datasets with augmentation
train_dataset = AugmentedDataset(piano_rolls[train_idx], emotion_labels[train_idx], augment=True)
val_dataset = AugmentedDataset(piano_rolls[val_idx], emotion_labels[val_idx], augment=False)
test_dataset = AugmentedDataset(piano_rolls[test_idx], emotion_labels[test_idx], augment=False)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)

print(f"Train: {len(train_dataset)}, Val: {len(val_dataset)}, Test: {len(test_dataset)}")

# Train ensemble
models = []
model_names = ['CNN', 'LSTM', 'ResNet']

for i, (ModelClass, name) in enumerate([(CNNClassifier, 'CNN'),
                                          (LSTMClassifier, 'LSTM'),
                                          (ResNetClassifier, 'ResNet')]):
    print(f"\n{'='*70}")
    print(f"Training Model {i+1}/3: {name}")
    print(f"{'='*70}")

    model = ModelClass().to(device)
    val_acc = train_model(model, train_loader, val_loader, epochs=80, lr=1e-3)

    print(f"✓ {name} Best Val Acc: {val_acc:.2f}%")

    torch.save(model.state_dict(), f'model_{name.lower()}.pth')
    models.append(model)

# ============================================================================
# ENSEMBLE EVALUATION
# ============================================================================

print("\n" + "=" * 70)
print("ENSEMBLE EVALUATION ON TEST SET")
print("=" * 70)

# Evaluate ensemble
all_preds_ensemble = []
all_labels = []

for model in models:
    model.eval()

with torch.no_grad():
    for data, labels in test_loader:
        data = data.to(device)

        # Get predictions from all models
        ensemble_logits = torch.zeros(data.size(0), 4).to(device)
        for model in models:
            logits = model(data)
            ensemble_logits += F.softmax(logits, dim=1)

        # Average predictions
        _, predicted = ensemble_logits.max(1)

        all_preds_ensemble.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())

all_preds_ensemble = np.array(all_preds_ensemble)
all_labels = np.array(all_labels)

test_acc = accuracy_score(all_labels, all_preds_ensemble)
print(f"\n✓ ENSEMBLE TEST ACCURACY: {test_acc*100:.2f}%")

# Individual model accuracies
print("\nIndividual Model Accuracies:")
for i, (model, name) in enumerate(zip(models, model_names)):
    model.eval()
    preds = []
    with torch.no_grad():
        for data, _ in test_loader:
            data = data.to(device)
            outputs = model(data)
            _, predicted = outputs.max(1)
            preds.extend(predicted.cpu().numpy())

    acc = accuracy_score(all_labels, np.array(preds))
    print(f"  {name}: {acc*100:.2f}%")

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds_ensemble)
emotion_names = ['Q1: Happy', 'Q2: Angry', 'Q3: Sad', 'Q4: Peaceful']

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=emotion_names, yticklabels=emotion_names)
plt.title(f'Ensemble Test Accuracy: {test_acc*100:.1f}%',
          fontsize=14, fontweight='bold', pad=20)
plt.ylabel('True Emotion', fontsize=12)
plt.xlabel('Predicted Emotion', fontsize=12)
plt.tight_layout()
plt.savefig('ensemble_confusion_matrix.png', dpi=100, bbox_inches='tight')
plt.show()

# Classification Report
print("\n" + "=" * 70)
print("CLASSIFICATION REPORT")
print("=" * 70)
print(classification_report(all_labels, all_preds_ensemble, target_names=emotion_names, digits=3))

# Per-class metrics
precision, recall, f1, support = precision_recall_fscore_support(all_labels, all_preds_ensemble)

print("\nPer-Class Metrics:")
for i, name in enumerate(emotion_names):
    print(f"{name:15s}: Precision={precision[i]:.3f}, Recall={recall[i]:.3f}, F1={f1[i]:.3f}")

# Save ensemble
print("\n" + "=" * 70)
print("SAVING ENSEMBLE")
print("=" * 70)

torch.save({
    'cnn': models[0].state_dict(),
    'lstm': models[1].state_dict(),
    'resnet': models[2].state_dict(),
    'test_acc': test_acc
}, 'ensemble_classifier.pth')

print(f"✓ Saved: ensemble_classifier.pth")
print(f"\nFinal Ensemble Test Accuracy: {test_acc*100:.2f}%")

# Create simple wrapper for inference
class EnsembleClassifier(nn.Module):
    def __init__(self):
        super(EnsembleClassifier, self).__init__()
        self.cnn = CNNClassifier()
        self.lstm = LSTMClassifier()
        self.resnet = ResNetClassifier()

    def forward(self, x):
        logits1 = F.softmax(self.cnn(x), dim=1)
        logits2 = F.softmax(self.lstm(x), dim=1)
        logits3 = F.softmax(self.resnet(x), dim=1)
        return (logits1 + logits2 + logits3) / 3

ensemble = EnsembleClassifier().to(device)
ensemble.cnn.load_state_dict(models[0].state_dict())
ensemble.lstm.load_state_dict(models[1].state_dict())
ensemble.resnet.load_state_dict(models[2].state_dict())

torch.save(ensemble.state_dict(), 'best_emotion_classifier.pth')
print("✓ Saved unified ensemble: best_emotion_classifier.pth")

print("\n" + "=" * 70)
print("✓ COMPLETE!")
print("=" * 70)
print(f"Ensemble achieves {test_acc*100:.2f}% on test set")
print("This model uses 3 different architectures voting together")
```

```python
# Step 6: Generate Emotion-Conditioned Music & Comprehensive Evaluation
# Updated to use the improved classifier
# Run this after training the improved classifier

import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import pretty_midi
import os
from collections import Counter
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import pickle

# Install audio dependencies
print("Installing audio conversion dependencies...")
!apt-get install -y fluidsynth > /dev/null 2>&1
!pip install -q midi2audio

print("=" * 60)
print("STEP 6: EMOTION-CONDITIONED MUSIC GENERATION & EVALUATION")
print("=" * 60)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nDevice: {device}")

# ============================================================
# DEFINE MODEL ARCHITECTURES
# ============================================================

class EmotionConditionedVAE(nn.Module):
    def __init__(self, input_dim=64*128, hidden_dim=1024, latent_dim=512,
                 num_emotions=4, emotion_embed_dim=64):
        super(EmotionConditionedVAE, self).__init__()
        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.emotion_embed_dim = emotion_embed_dim

        self.emotion_embedding = nn.Embedding(num_emotions, emotion_embed_dim)

        self.encoder = nn.Sequential(
            nn.Linear(input_dim + emotion_embed_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2)
        )

        self.fc_mu = nn.Linear(hidden_dim // 2, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim // 2, latent_dim)

        self.decoder = nn.Sequential(
            nn.Linear(latent_dim + emotion_embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 2, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()
        )

    def decode(self, z, emotion):
        emotion_emb = self.emotion_embedding(emotion)
        z_cond = torch.cat([z, emotion_emb], dim=1)
        x = self.decoder(z_cond)
        return x.view(-1, 64, 128)

class EmotionConditionedGenerator(nn.Module):
    def __init__(self, noise_dim=128, latent_dim=512, hidden_dim=512,
                 num_emotions=4, emotion_embed_dim=64):
        super(EmotionConditionedGenerator, self).__init__()

        self.emotion_embedding = nn.Embedding(num_emotions, emotion_embed_dim)

        self.model = nn.Sequential(
            nn.Linear(noise_dim + emotion_embed_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.BatchNorm1d(hidden_dim),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.BatchNorm1d(hidden_dim),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.BatchNorm1d(hidden_dim),
            nn.Linear(hidden_dim, latent_dim)
        )

    def forward(self, noise, emotion):
        emotion_emb = self.emotion_embedding(emotion)
        x = torch.cat([noise, emotion_emb], dim=1)
        return self.model(x)

class CNNClassifier(nn.Module):
    """Pure CNN architecture"""
    def __init__(self, num_emotions=4):
        super(CNNClassifier, self).__init__()

        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=(3, 7), padding=(1, 3)),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d((2, 2)),
            nn.Dropout2d(0.3),

            nn.Conv2d(32, 64, kernel_size=(3, 7), padding=(1, 3)),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d((2, 2)),
            nn.Dropout2d(0.3),

            nn.Conv2d(64, 128, kernel_size=(3, 5), padding=(1, 2)),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 8))
        )

        self.fc = nn.Sequential(
            nn.Linear(128*4*8, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.BatchNorm1d(256),
            nn.Linear(256, num_emotions)
        )

    def forward(self, x):
        x = self.conv(x.unsqueeze(1))
        x = x.flatten(1)
        return self.fc(x)

class LSTMClassifier(nn.Module):
    """LSTM-based architecture"""
    def __init__(self, num_emotions=4):
        super(LSTMClassifier, self).__init__()

        self.lstm = nn.LSTM(
            input_size=128,
            hidden_size=256,
            num_layers=2,
            batch_first=True,
            bidirectional=True,
            dropout=0.3
        )

        self.attention = nn.Sequential(
            nn.Linear(512, 256),
            nn.Tanh(),
            nn.Linear(256, 1)
        )

        self.fc = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.BatchNorm1d(256),
            nn.Linear(256, num_emotions)
        )

    def forward(self, x):
        # x: (batch, 64, 128)
        lstm_out, _ = self.lstm(x)  # (batch, 64, 512)

        # Attention
        attn_weights = self.attention(lstm_out)  # (batch, 64, 1)
        attn_weights = torch.nn.functional.softmax(attn_weights, dim=1)

        # Weighted sum
        context = torch.sum(lstm_out * attn_weights, dim=1)  # (batch, 512)

        return self.fc(context)

class ResNetClassifier(nn.Module):
    """ResNet-style architecture"""
    def __init__(self, num_emotions=4):
        super(ResNetClassifier, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, kernel_size=7, padding=3)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(2)

        # Residual blocks
        self.res1 = self._make_residual_block(32, 64)
        self.res2 = self._make_residual_block(64, 128)
        self.res3 = self._make_residual_block(128, 256)

        self.global_pool = nn.AdaptiveAvgPool2d((1, 1))

        self.fc = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_emotions)
        )

    def _make_residual_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(),
            nn.Conv2d(out_channels, out_channels, 3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.MaxPool2d(2)
        )

    def forward(self, x):
        x = x.unsqueeze(1)
        x = torch.nn.functional.relu(self.bn1(self.conv1(x)))
        x = self.pool1(x)

        x = self.res1(x)
        x = self.res2(x)
        x = self.res3(x)

        x = self.global_pool(x)
        x = x.flatten(1)
        return self.fc(x)

class EnsembleClassifier(nn.Module):
    """Ensemble of CNN, LSTM, and ResNet"""
    def __init__(self):
        super(EnsembleClassifier, self).__init__()
        self.cnn = CNNClassifier()
        self.lstm = LSTMClassifier()
        self.resnet = ResNetClassifier()

    def forward(self, x):
        logits1 = torch.nn.functional.softmax(self.cnn(x), dim=1)
        logits2 = torch.nn.functional.softmax(self.lstm(x), dim=1)
        logits3 = torch.nn.functional.softmax(self.resnet(x), dim=1)
        return (logits1 + logits2 + logits3) / 3

# ============================================================
# FEATURE EXTRACTION - NOT NEEDED FOR ENSEMBLE (uses piano rolls directly)
# ============================================================

# Remove feature extraction functions since ensemble takes piano rolls directly

# ============================================================
# LOAD TRAINED MODELS
# ============================================================

print("\nLoading trained models...")

vae = EmotionConditionedVAE().to(device)
vae.load_state_dict(torch.load('emotion_vae_model.pth', map_location=device))
vae.eval()
print("✓ VAE loaded")

generator = EmotionConditionedGenerator().to(device)
generator.load_state_dict(torch.load('emotion_generator.pth', map_location=device))
generator.eval()
print("✓ Generator loaded")

# Load ENSEMBLE classifier
ensemble_checkpoint = torch.load('best_emotion_classifier.pth', map_location=device)
classifier = EnsembleClassifier().to(device)
classifier.load_state_dict(ensemble_checkpoint)
classifier.eval()
print(f"✓ Ensemble Classifier loaded (3 models: CNN + LSTM + ResNet)")

# Load latent normalization stats
latent_stats = torch.load('emotion_latent_stats.pth', map_location=device)
latent_mean = latent_stats['mean'].to(device)
latent_std = latent_stats['std'].to(device)
print("✓ Latent normalization stats loaded")

# ============================================================
# GENERATION FUNCTION
# ============================================================

def generate_music(emotion_class, num_samples=1, temperature=1.0):
    """Generate music with specific emotion"""
    with torch.no_grad():
        noise = torch.randn(num_samples, 128).to(device) * temperature
        emotions = torch.LongTensor([emotion_class] * num_samples).to(device)

        latent_normalized = generator(noise, emotions)
        latent = latent_normalized * latent_std + latent_mean
        piano_rolls = vae.decode(latent, emotions)

        return piano_rolls.cpu().numpy()

# ============================================================
# PIANO ROLL TO MIDI
# ============================================================

def piano_roll_to_midi(piano_roll, filename='generated.mid', threshold=0.3, tempo=120):
    """Convert piano roll to MIDI file"""
    midi = pretty_midi.PrettyMIDI(initial_tempo=tempo)
    piano_program = pretty_midi.instrument_name_to_program('Acoustic Grand Piano')
    piano = pretty_midi.Instrument(program=piano_program)

    time_step = 60.0 / tempo / 4
    piano_roll_binary = (piano_roll > threshold).astype(int)

    active_notes = {}

    for time_idx in range(piano_roll_binary.shape[0]):
        for pitch in range(piano_roll_binary.shape[1]):
            is_active = piano_roll_binary[time_idx, pitch]

            if is_active and pitch not in active_notes:
                active_notes[pitch] = time_idx * time_step
            elif not is_active and pitch in active_notes:
                start_time = active_notes[pitch]
                end_time = time_idx * time_step
                note = pretty_midi.Note(
                    velocity=80, pitch=pitch,
                    start=start_time, end=end_time
                )
                piano.notes.append(note)
                del active_notes[pitch]

    final_time = piano_roll_binary.shape[0] * time_step
    for pitch, start_time in active_notes.items():
        note = pretty_midi.Note(
            velocity=80, pitch=pitch,
            start=start_time, end=final_time
        )
        piano.notes.append(note)

    midi.instruments.append(piano)
    midi.write(filename)
    return midi

# ============================================================
# GENERATE SAMPLES FOR EACH EMOTION
# ============================================================

print("\n" + "=" * 60)
print("GENERATING EMOTION-CONDITIONED MUSIC")
print("=" * 60)

os.makedirs('generated_emotion_music', exist_ok=True)

NUM_SAMPLES_PER_EMOTION = 50
TEMPERATURE = 1.0

emotion_names = {
    0: 'Q1_Happy',
    1: 'Q2_Angry',
    2: 'Q3_Sad',
    3: 'Q4_Peaceful'
}

all_generated_samples = {}
all_generated_labels = []

for emotion_class in range(4):
    print(f"\nGenerating {NUM_SAMPLES_PER_EMOTION} samples for {emotion_names[emotion_class]}...")

    piano_rolls = generate_music(emotion_class, NUM_SAMPLES_PER_EMOTION, TEMPERATURE)
    all_generated_samples[emotion_class] = piano_rolls
    all_generated_labels.extend([emotion_class] * NUM_SAMPLES_PER_EMOTION)

    # Save first 3 MIDI files
    for i in range(min(3, NUM_SAMPLES_PER_EMOTION)):
        filename = f'generated_emotion_music/{emotion_names[emotion_class]}_sample_{i+1}.mid'
        piano_roll_to_midi(piano_rolls[i], filename=filename, threshold=0.3, tempo=120)

    print(f"✓ Generated and saved samples for {emotion_names[emotion_class]}")

print(f"\n✓ Total generated: {len(all_generated_labels)} samples")

# ============================================================
# CLASSIFY GENERATED SAMPLES
# ============================================================

print("\n" + "=" * 60)
print("CLASSIFYING GENERATED SAMPLES WITH ENSEMBLE")
print("=" * 60)

all_predictions = []

classifier.eval()
with torch.no_grad():
    for emotion_class in range(4):
        piano_rolls = all_generated_samples[emotion_class]
        piano_rolls_tensor = torch.FloatTensor(piano_rolls).to(device)

        # Classify with ensemble (just pass piano rolls directly)
        outputs = classifier(piano_rolls_tensor)
        _, preds = outputs.max(1)
        all_predictions.extend(preds.cpu().numpy())

all_predictions = np.array(all_predictions)
all_generated_labels = np.array(all_generated_labels)

print("✓ Classification complete")

# ============================================================
# EVALUATION METRICS
# ============================================================

print("\n" + "=" * 60)
print("EVALUATION: EMOTION CONDITIONING ACCURACY")
print("=" * 60)

conditioning_accuracy = accuracy_score(all_generated_labels, all_predictions)
print(f"\n✓ Overall Emotion Conditioning Accuracy: {conditioning_accuracy*100:.2f}%")
print(f"  (How often generated music matches intended emotion)")

# Confusion matrix
cm = confusion_matrix(all_generated_labels, all_predictions)
emotion_display_names = ['Q1: Happy', 'Q2: Angry', 'Q3: Sad', 'Q4: Peaceful']

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=emotion_display_names, yticklabels=emotion_display_names)
plt.title(f'Generated Music - Conditioning Accuracy: {conditioning_accuracy*100:.1f}%',
          fontsize=14, fontweight='bold', pad=20)
plt.ylabel('Intended Emotion', fontsize=12, fontweight='bold')
plt.xlabel('Classified Emotion', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('generated_music_confusion_matrix.png', dpi=100, bbox_inches='tight')
plt.show()

# Per-emotion accuracy
print("\n" + "=" * 60)
print("PER-EMOTION CONDITIONING ACCURACY")
print("=" * 60)

for emotion_class in range(4):
    mask = all_generated_labels == emotion_class
    emotion_preds = all_predictions[mask]
    emotion_labels = all_generated_labels[mask]

    accuracy = accuracy_score(emotion_labels, emotion_preds)
    total = len(emotion_labels)
    correct = (emotion_preds == emotion_labels).sum()

    print(f"\n{emotion_display_names[emotion_class]}:")
    print(f"  Accuracy: {accuracy*100:.2f}% ({correct}/{total})")

    if correct < total:
        misclass_counts = Counter(emotion_preds[emotion_preds != emotion_class])
        print(f"  Misclassified as:")
        for misclass_emotion, count in misclass_counts.items():
            print(f"    {emotion_display_names[misclass_emotion]}: {count}")

print("\n" + "=" * 60)
print(classification_report(all_generated_labels, all_predictions,
                           target_names=emotion_display_names, digits=3))

print("\n" + "=" * 60)
print("✓✓✓ COMPLETE! ✓✓✓")
print("=" * 60)
print(f"\nFinal Results:")
print(f"  Ensemble Classifier (CNN + LSTM + ResNet)")
print(f"  Generation Conditioning Accuracy: {conditioning_accuracy*100:.2f}%")
print(f"  Generated Music: generated_emotion_music/")
print(f"\nNote: Ensemble uses 3 models voting together for robustness")
```

```python
# Evaluation Metrics for Generated Music
# Add this after Step 5 generation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance
from scipy.special import kl_div

print("=" * 60)
print("COMPREHENSIVE EVALUATION METRICS")
print("=" * 60)

# Load data
train_piano_rolls = np.load('piano_rolls.npy')
generated_piano_rolls = generated_piano_rolls  # From Step 5

THRESHOLD = 0.3  # Note activation threshold

print(f"\nDataset sizes:")
print(f"  Training samples: {len(train_piano_rolls)}")
print(f"  Generated samples: {len(generated_piano_rolls)}")

# ============================================
# 1. NOTE DENSITY METRICS
# ============================================
print("\n" + "=" * 60)
print("1. NOTE DENSITY ANALYSIS")
print("=" * 60)

train_binary = (train_piano_rolls > 0).astype(float)
gen_binary = (generated_piano_rolls > THRESHOLD).astype(float)

train_density = train_binary.mean()
gen_density = gen_binary.mean()

print(f"\nOverall note density (% active):")
print(f"  Training: {train_density*100:.2f}%")
print(f"  Generated: {gen_density*100:.2f}%")
print(f"  Difference: {abs(train_density - gen_density)*100:.2f}%")

# Density over time
train_density_per_step = train_binary.mean(axis=(0, 2))
gen_density_per_step = gen_binary.mean(axis=(0, 2))

plt.figure(figsize=(10, 5))
plt.plot(train_density_per_step, label='Training', linewidth=2)
plt.plot(gen_density_per_step, label='Generated', linewidth=2)
plt.xlabel('Time Step', fontsize=12)
plt.ylabel('Note Density', fontsize=12)
plt.title('Note Density Over Time', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('metric_density_over_time.png', dpi=100)
plt.show()

# ============================================
# 2. NOTES PER SAMPLE
# ============================================
print("\n" + "=" * 60)
print("2. NOTES PER SAMPLE STATISTICS")
print("=" * 60)

train_notes = train_binary.sum(axis=(1, 2))
gen_notes = gen_binary.sum(axis=(1, 2))

print(f"\nNotes per sample:")
print(f"  Training: {train_notes.mean():.1f} ± {train_notes.std():.1f}")
print(f"  Generated: {gen_notes.mean():.1f} ± {gen_notes.std():.1f}")
print(f"  Min/Max (train): {train_notes.min():.0f} / {train_notes.max():.0f}")
print(f"  Min/Max (gen): {gen_notes.min():.0f} / {gen_notes.max():.0f}")

# Histogram comparison
plt.figure(figsize=(10, 5))
plt.hist(train_notes, bins=30, alpha=0.6, label='Training', density=True)
plt.hist(gen_notes, bins=30, alpha=0.6, label='Generated', density=True)
plt.xlabel('Number of Notes', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Distribution of Notes Per Sample', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('metric_notes_distribution.png', dpi=100)
plt.show()

# Wasserstein distance (Earth Mover's Distance)
wd = wasserstein_distance(train_notes, gen_notes)
print(f"\nWasserstein Distance: {wd:.2f}")

# ============================================
# 3. PITCH RANGE ANALYSIS
# ============================================
print("\n" + "=" * 60)
print("3. PITCH RANGE ANALYSIS")
print("=" * 60)

train_pitch_usage = train_binary.sum(axis=(0, 1))  # Sum over samples and time
gen_pitch_usage = gen_binary.sum(axis=(0, 1))

train_active_pitches = np.where(train_pitch_usage > 0)[0]
gen_active_pitches = np.where(gen_pitch_usage > 0)[0]

print(f"\nActive pitches:")
print(f"  Training: {len(train_active_pitches)}/128")
print(f"  Generated: {len(gen_active_pitches)}/128")

if len(train_active_pitches) > 0:
    print(f"\nPitch range (MIDI):")
    print(f"  Training: {train_active_pitches.min()} - {train_active_pitches.max()} (span: {train_active_pitches.max() - train_active_pitches.min()})")
if len(gen_active_pitches) > 0:
    print(f"  Generated: {gen_active_pitches.min()} - {gen_active_pitches.max()} (span: {gen_active_pitches.max() - gen_active_pitches.min()})")

# Pitch usage histogram
plt.figure(figsize=(12, 5))
plt.bar(range(128), train_pitch_usage, alpha=0.6, label='Training')
plt.bar(range(128), gen_pitch_usage, alpha=0.6, label='Generated')
plt.xlabel('MIDI Pitch', fontsize=12)
plt.ylabel('Total Usage Count', fontsize=12)
plt.title('Pitch Usage Distribution', fontsize=14, fontweight='bold')
plt.xlim([20, 110])  # Piano range
plt.legend()
plt.grid(alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('metric_pitch_distribution.png', dpi=100)
plt.show()

# ============================================
# 4. POLYPHONY (NOTES PER TIMESTEP)
# ============================================
print("\n" + "=" * 60)
print("4. POLYPHONY ANALYSIS")
print("=" * 60)

train_polyphony = train_binary.sum(axis=2)  # Notes per timestep
gen_polyphony = gen_binary.sum(axis=2)

train_poly_mean = train_polyphony.mean(axis=0)  # Average per timestep
gen_poly_mean = gen_polyphony.mean(axis=0)

print(f"\nAverage polyphony (simultaneous notes):")
print(f"  Training: {train_polyphony.mean():.2f}")
print(f"  Generated: {gen_polyphony.mean():.2f}")

plt.figure(figsize=(10, 5))
plt.plot(train_poly_mean, label='Training', linewidth=2)
plt.plot(gen_poly_mean, label='Generated', linewidth=2)
plt.xlabel('Time Step', fontsize=12)
plt.ylabel('Average Simultaneous Notes', fontsize=12)
plt.title('Polyphony Over Time', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('metric_polyphony.png', dpi=100)
plt.show()

# ============================================
# 5. EMPTY SAMPLES DETECTION
# ============================================
print("\n" + "=" * 60)
print("5. EMPTY SAMPLES DETECTION")
print("=" * 60)

train_empty = (train_notes == 0).sum()
gen_empty = (gen_notes == 0).sum()

print(f"\nEmpty samples (no notes):")
print(f"  Training: {train_empty}/{len(train_piano_rolls)} ({train_empty/len(train_piano_rolls)*100:.1f}%)")
print(f"  Generated: {gen_empty}/{len(generated_piano_rolls)} ({gen_empty/len(generated_piano_rolls)*100:.1f}%)")

# ============================================
# 6. PITCH CLASS DISTRIBUTION (KL DIVERGENCE)
# ============================================
print("\n" + "=" * 60)
print("6. PITCH CLASS DISTRIBUTION SIMILARITY")
print("=" * 60)

# Normalize to probability distributions
train_pitch_dist = train_pitch_usage / (train_pitch_usage.sum() + 1e-10)
gen_pitch_dist = gen_pitch_usage / (gen_pitch_usage.sum() + 1e-10)

# KL Divergence (lower is better)
kl_divergence = np.sum(kl_div(train_pitch_dist + 1e-10, gen_pitch_dist + 1e-10))
print(f"\nKL Divergence (pitch distributions): {kl_divergence:.4f}")
print("  (Lower is better; 0 = identical distributions)")

# ============================================
# 7. TEMPORAL CONTINUITY
# ============================================
print("\n" + "=" * 60)
print("7. TEMPORAL CONTINUITY ANALYSIS")
print("=" * 60)

# Measure note onset density (how many notes start at each timestep)
def calculate_onsets(piano_rolls_binary):
    """Calculate note onsets (transitions from 0 to 1)"""
    onsets = np.zeros_like(piano_rolls_binary)
    onsets[:, 0, :] = piano_rolls_binary[:, 0, :]  # First timestep
    onsets[:, 1:, :] = np.diff(piano_rolls_binary, axis=1) > 0
    return onsets

train_onsets = calculate_onsets(train_binary)
gen_onsets = calculate_onsets(gen_binary)

train_onset_density = train_onsets.sum(axis=(0, 2)) / len(train_piano_rolls)
gen_onset_density = gen_onsets.sum(axis=(0, 2)) / len(generated_piano_rolls)

plt.figure(figsize=(10, 5))
plt.plot(train_onset_density, label='Training', linewidth=2)
plt.plot(gen_onset_density, label='Generated', linewidth=2)
plt.xlabel('Time Step', fontsize=12)
plt.ylabel('Average Note Onsets', fontsize=12)
plt.title('Note Onset Density (Rhythm Activity)', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('metric_onset_density.png', dpi=100)
plt.show()

print(f"\nAverage onsets per timestep:")
print(f"  Training: {train_onsets.sum(axis=2).mean():.2f}")
print(f"  Generated: {gen_onsets.sum(axis=2).mean():.2f}")

# ============================================
# 8. SUMMARY METRICS TABLE
# ============================================
print("\n" + "=" * 60)
print("SUMMARY METRICS TABLE")
print("=" * 60)

summary = f"""
╔═══════════════════════════════════════╦═══════════╦═══════════╗
║ Metric                                ║ Training  ║ Generated ║
╠═══════════════════════════════════════╬═══════════╬═══════════╣
║ Note Density (%)                      ║ {train_density*100:>8.2f}% ║ {gen_density*100:>8.2f}% ║
║ Avg Notes per Sample                  ║ {train_notes.mean():>9.1f} ║ {gen_notes.mean():>9.1f} ║
║ Avg Polyphony (simultaneous notes)    ║ {train_polyphony.mean():>9.2f} ║ {gen_polyphony.mean():>9.2f} ║
║ Active Pitches                        ║ {len(train_active_pitches):>9}/128 ║ {len(gen_active_pitches):>9}/128 ║
║ Empty Samples                         ║ {train_empty:>9} ║ {gen_empty:>9} ║
║ Avg Onsets per Timestep               ║ {train_onsets.sum(axis=2).mean():>9.2f} ║ {gen_onsets.sum(axis=2).mean():>9.2f} ║
╚═══════════════════════════════════════╩═══════════╩═══════════╝

Additional Metrics:
  • Wasserstein Distance (notes): {wd:.2f}
  • KL Divergence (pitch dist): {kl_divergence:.4f}
"""

print(summary)

# Save metrics to file
with open('evaluation_metrics.txt', 'w') as f:
    f.write(summary)

print("\n✓ All metrics calculated and saved!")
print("✓ Visualizations saved to metric_*.png files")
print("✓ Summary saved to evaluation_metrics.txt")
```
